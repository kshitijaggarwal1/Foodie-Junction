```python
def function6part1(driver):
    driver.find_element_by_xpath('//ul[@class="k9GMp "]/li[2]/a').click()
    time.sleep(3)
    count = 1
    followers = []
    followers_panel = driver.find_element_by_class_name('isgrP')
    res = False
    while count <= 500:
        try:
            link1 = '//div[@class="PZuss"]/li[' + str(count) + ']/div/div[2]/div/div/div/a'
            link2 = '//div[@class="PZuss"]/li[' + str(count) + ']/div/div/div[2]/div/a'
            if res == False:
                username = driver.find_element_by_xpath(link1).text
            else:
                username = driver.find_element_by_xpath(link2).text
            followers.append(username)
            count+=1
        except NoSuchElementException:
            driver.execute_script('arguments[0].scrollBy(0,1000);',followers_panel)
            time.sleep(2)
            res = True
    print('List of usernames of top 500 followers: ')
    print(followers)
    
    driver.find_element_by_xpath('//div[@class="eiUFA"]/div[2]/button').click()
    time.sleep(1)
    
#main
time.sleep(2)

#for foodtalkindia
search = driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/input')
search.send_keys('foodtalkindia')
time.sleep(3)
driver.find_element_by_xpath('//*[@class="drKGC"]/div/a').click()
time.sleep(2)
print('for foodtalkindia: ')
function6part1(driver)

#for sodelhi
search = driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/input')
search.send_keys('sodelhi')
time.sleep(3)
driver.find_element_by_xpath('//*[@class="drKGC"]/div/a').click()
time.sleep(2)
print('for sodelhi: ')
function6part1(driver)
```

```python
def function6part2(driver):
    
    final_out = False
    followers = []
    while True:
        try:
            driver.find_element_by_class_name('tc8A9').click()
            time.sleep(1)
            driver.find_element_by_xpath('//div[@class="SfPzb"]/a').click()
            time.sleep(3)
            followers_panel = driver.find_element_by_class_name('isgrP')
            res = False
            i = 1
            while True:
                try:
                    link1 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div[2]/div/div/div/a'
                    link2 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div/div[2]/div/a'
                    button_link1 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div[3]/button'
                    button_link2 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div[2]/button'
                    if res == False:
                        button_text = driver.find_element_by_xpath(button_link1).text
                    else:
                        button_text = driver.find_element_by_xpath(button_link2).text
                    if button_text == 'Follow':
                        final_out = True
                        break

                    if res == False:
                        username = driver.find_element_by_xpath(link1).text
                    else:
                        username = driver.find_element_by_xpath(link2).text
                    followers.append(username)
                    i+=1
                except NoSuchElementException:
                    driver.execute_script('arguments[0].scrollBy(0,1000);',followers_panel)
                    time.sleep(2)
                    res = True
            if final_out == True:
                break
        except NoSuchElementException:
            print('No common followers')
            break
    driver.find_element_by_xpath('//div[@class="eiUFA"]/div[2]/button').click()
    time.sleep(2)
    return followers

def followers(driver):
    followers_list = []
    total = driver.find_element_by_xpath('//ul[@class="k9GMp "]/li[2]/a/span').text
    driver.find_element_by_xpath('//ul[@class="k9GMp "]/li[2]/a').click()
    followers_panel = driver.find_element_by_class_name('isgrP')
    i = 1
    res = False
    
    while i <= int(total):
        try:
            link1 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div[2]/div/div/div/a'
            link2 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div/div[2]/div/a'
            if res == False:
                username = driver.find_element_by_xpath(link1).text
            else:
                username = driver.find_element_by_xpath(link2).text
            followers_list.append(username)
            i+=1
        except NoSuchElementException:
            driver.execute_script('arguments[0].scrollBy(0,1000);',followers_panel)
            time.sleep(4)
            res = True
    driver.find_element_by_xpath('//div[@class="eiUFA"]/div[2]/button').click()
    return followers_list

def following(driver):
    following_list = []
    total = driver.find_element_by_xpath('//ul[@class="k9GMp "]/li[3]/a/span').text
    driver.find_element_by_xpath('//ul[@class="k9GMp "]/li[3]/a').click()
    followers_panel = driver.find_element_by_class_name('isgrP')
    i = 1
    res = False
    
    while i <= int(total):
        try:
            link1 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div[2]/div/div/div/a'
            link2 = '//div[@class="PZuss"]/li[' + str(i) + ']/div/div/div[2]/div/a'
            if res == False:
                username = driver.find_element_by_xpath(link1).text
            else:
                username = driver.find_element_by_xpath(link2).text
            following_list.append(username)
            i+=1
        except NoSuchElementException:
            driver.execute_script('arguments[0].scrollBy(0,1000);',followers_panel)
            time.sleep(4)
            res = True
    driver.find_element_by_xpath('//div[@class="eiUFA"]/div[2]/button').click()
    return following_list
    
def helper(driver):
    driver.find_element_by_xpath('//a[contains(@class, "qNELH")]').click()
    time.sleep(3)
    my_list = []
    followers_list = []
    following_list = []
    #finding the people that I follow
    followers_list = followers(driver)
    #finding the people that follow me
    following_list = following(driver)
    my_list = list(set(following_list)-set(followers_list))
    return my_list

#main
#for finding the common followers of foodtalkindia and myself
search = driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/input')
search.send_keys('foodtalkindia')
time.sleep(3)
driver.find_element_by_xpath('//*[@class="drKGC"]/div/a').click()
time.sleep(2)
foodtalkindia_common_list = []
foodtalkindia_common_list = function6part2(driver)

#for finding people that I follow but they don't follow me
my_list = []
my_list = helper(driver)

#finding the common ones in both of them
ans = set(foodtalkindia_common_list).intersection(set(my_list))
for i in ans:
    print(i)
```
